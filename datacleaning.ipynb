```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('data.csv')

# Optional: Drop missing values in Maxpulse column
df = df[df['Maxpulse'].notna()]

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Maxpulse'], bins=20, color='lightcoral', edgecolor='black')

# Customize the plot
plt.title('Histogram of Maxpulse')
plt.xlabel('Maxpulse')
plt.ylabel('Frequency')
plt.grid(True)

# Show the plot
plt.show()
```
